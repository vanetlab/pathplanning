```python
# load from grid_traffic.csv
data= pd.read_csv('grid_E_traffic.csv')
Time=data["Timedata"]
dataAll=[]
filename="A"+str(1)+"_"+str(1)
for i in range(1,21):
    for j in range(1,21):
        filename="A"+str(i)+"_"+str(j)
        dataAll.append(data[filename].tolist())
        
```

```python
print(len(dataAll))
print(len(dataAll[0]))
print(len(dataAll[1]))
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import GRU
from keras.layers import Activation
from keras.layers import Dense
from keras.layers import Dropout
from keras.models import load_model
N=[]
A=[]
for i in range(0,400):
    A.append(5)
for i in range(0,400):
    N.append(sum(dataAll[i])/len(dataAll[i]))
ntable=[]
for i in range(1,401):
    tmpA=[number/N[i-1]+A[i-1] for number in dataAll[i-1]]
    ntable.append(tmpA)

        
        
    
```

```python
print(len(ntable[1]))
```

```python
n=300
x=8000
train_x=[]
train_y=[]
for m in range(0,400):
    xtmp=[]
    ytmp=[]
    for i in range(n,x):
        xtmp.append(ntable[m][i-n:i])
        ytmp.append(ntable[m][i])
    train_x.append(xtmp)
    train_y.append(ytmp)
    
```

```python
#print(len(train_x))
#print(len(train_x[0]))
#print(len(train_x[0][0]))
#print(len(train_y[0]))
print(len(train_x))
print(len(train_x[0]))
print(len(train_x[0][0]))

```

```python
for m in range(195,205):
    train_tmp_x, train_tmp_y=np.array(train_x[m]),np.array(train_y[m])
    train_tmp_x=train_tmp_x.reshape(train_tmp_x.shape[0],train_tmp_x.shape[1],1)
    print(train_tmp_x.shape)
    print(train_tmp_y.shape)
    #training for tmp
    regressor=Sequential()
    regressor.add(LSTM(units=n,return_sequences=True,input_shape=(train_tmp_x.shape[1],train_tmp_x.shape[2])))
    regressor.add(Dropout(0.05))
    regressor.add(LSTM(units=n))
    regressor.add(Dropout(0.05))
    regressor.add(Dense(units=1))
    regressor.compile(optimizer='adam',loss='mean_squared_error')
    regressor.fit(train_tmp_x,train_tmp_y,epochs=1,batch_size=30)
    filename=str(m)+"model.h5"
    regressor.save(filename)
    model=load_model(filename)
    predicted_values=model.predict(train_tmp_x)
    plt.figure()
    plt.plot(train_tmp_y,label="training data")
    plt.plot(predicted_values,label="prediction")
    temp1=[]
    temp2=[]
    temp2 = [(number-A[m])*N[m] for number in train_tmp_y]
    predicted_values=model.predict(train_tmp_x)
    temp1 = [(number-A[m])*N[m] for number in predicted_values]
    temp1=[np.round(number) for number in temp1]
    plt.figure(figsize=(10,5))
    plt.plot(temp2,label="training data")
    plt.plot(temp1,label="prediction")
    plt.legend()
    plt.xlabel("time")
    plt.ylabel("A density")
```
